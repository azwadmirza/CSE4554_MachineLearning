### Given a matrix A, calculate A$^{T}$A where T denotes the transpose.

```python
A=np.array([[1,2,1],[3,0,1],[0,2,4]])
np.matmul(np.transpose(A),A)
```

### Multiply a 4x3 matrix B with a 3x2 matrix A. Confirm the dimensions of the resulting matrix.

Mathematically the result of multiplying 4x3 matrix with 3x2 matrix should be:

4 x ~3~

~3~ x 2

--------
4 x 2


```python
A=np.random.randint(10,size=(4,3))
B=np.random.randint(10,size=(3,2))
C=np.matmul(A,B)
print("Matrix is: ",C)
print("Shape is {0}x{1}".format(C.shape[0],C.shape[1]))
```

### Given two matrices A and B, show that in general A · B ̸= B · A.

```python
A=np.random.randint(10,size=(3,3))
B=np.random.randint(10,size=(3,3))
print("A = ",A)
print("B = ",B)
print("A.B = ",np.matmul(A,B))
print("B.A = ",np.matmul(B,A))
```
